## Data Loading & Exploration
Load and have a look at the provided data.

```python
# Load the dataset
df = pd.read_csv('./kyphosis.csv', low_memory=False)
df.head(10)
```

```python
print('Number of samples: {}, number of attributes: {}'.format(df.shape[0], df.shape[1]))
```

```python
# Get info on the data types
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(data=df, hue='Kyphosis', palette='Set1')
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(
    df.drop('Kyphosis', axis=1), df['Kyphosis'], test_size=0.2, random_state=42)
```

```python
[x.shape for x in (train_data, test_data, train_labels, test_labels)]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(train_data, train_labels)
```

```python
test_preds = dtree.predict(test_data)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(test_labels, test_preds))
```

```python
print(confusion_matrix(test_labels, test_preds))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(train_data, train_labels)
```

```python
test_preds = rfc.predict(test_data)
```

```python
print(classification_report(test_labels, test_preds))
```

```python
print(confusion_matrix(test_labels, test_preds))
```
