### ========== Question 1.7 --- [8 marks] ==========

We will now investigate using dimensionality reduction to try and improve the quality of the fit. Use the sklearn implementation of [PCA](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) and its method `fit_transform()` to create a reduced reduced dataset for `n_components` = [1,2,3,5,10,20,50,100,500,1000] i.e. create datasets that are of shape `(N, d)` for `d` in `n_components`. Fit k-means to each reduced dataset and report the `inertia` and `adjusted_rand_score` for each iteration.

Plot `adjusted_rand_score` against number of principal components (label graph). Use a log scale on the x axis. Below the graph:
1. describe what it shows
1. explain why we cannot use inertia to choose the best number of principal components
1. explain why dimensionality reduction could help k-means perform better

```python
n_components = [1, 2, 3, 5, 10, 20, 50, 100, 500, 1000]
for d in n_components:
    pca = PCA(n_components=d).fit_transform(X)
    print ("The shape of X after PCA with {} components: {}".format(d, pca.shape))
    kmeans = KMeans(n_clusters = 5, random_state = 1000).fit(pca)
    print ("k-means classifier inertia: {}, adjusted random score: {}".format(kmeans.inertia_, adjusted_rand_score(y, kmeans.labels_)))
```

* *1. It can be observed that the inertia of the kmeans classifier gradually increases with increasing amount of feature dimensions. The adjusted random score varies slightly, with the smallest value obtained with 500 principal components, and the highest with 2.*
* *2. We cannot choose the best model using inertia because this metric is based on distance which generally increases with higher data dimensions (or amount of features for each data point). Hence, despite the fact that model might show better adjusted random score with more features (e.g., 100 vs 500 in the present example), the inertia will show an increase too. Essentially, the inertia for comparison can be used only in cases where the models use the same amount of dimensions to describe the data.*
* *3. When dimensionality of the data is big, there are two problems that may arise: high computational cost of training the models, and overfitting. Dimensionality reduction allows to cope with both of these problems, resulting in models that generalize better and are faster to train. To put it in other terms, the dimensionality reduction may allow to get rid of specificities of the data, and instead produce a more general representation of each class, often more distinguishable. Finally, reducing dimensionality allows to visualize the data, which is crucial for understanding the relations between datapoints and their features.*

Create a new `PCA` object with `n_components = digits.data.shape[1]`. Plot the explained variance **ratio** against the number of components. You may find [this example](http://scikit-learn.org/stable/auto_examples/plot_digits_pipe.html#sphx-glr-auto-examples-plot-digits-pipe-py) quite handy...

Find the point where 95% of the variance has been explained. Use `plt.vlines()` to add a dotted verical line to the graph at that point and use `plt.annotate()` to label the number of eigenvectors used to explain that variance.

Below the plot, explain what you observe. What does this suggest about the data?

```python
pca = PCA(n_components=digits.data.shape[1]).fit(digits.data)
plt.figure(1, figsize=(10, 6))
plt.clf()
ratio = pca.explained_variance_ratio_
plt.plot(ratio, linewidth=2)
plt.axis('tight')
plt.xlabel('Number of components')
plt.ylabel('Explained variance ratio')

ratio_sum = 0
n_eigen = 0
result_x = 0
for r in ratio:
    n_eigen += 1
    ratio_sum += r
    if (ratio_sum >= 0.95): break

plt.vlines(n_eigen, -1, 1, color='k',linestyles='dotted')
annotation = "Amount of components that explain 95% of the variance is  " + str(n_eigen)
plt.annotate(annotation, xy=(n_eigen, 0.01), xytext=(n_eigen-10, 0.06),
            arrowprops=dict(facecolor='black', width=0.4, headwidth=7))

plt.show()
```

*It can be observed that the ratio of explained variance drops down severely with increasing amount of components at the start, but the drop becomes smaller after each of such additions. Essentially, more components is better, but just to some extent. Here, we can see that more than 95% of all variance can be explained by less than a half of available components (29), and addition of more would not significantly reduce the explained variance. This also suggest that full dimensionality of data is not really necessary, some of the information stored in all of those dimensions can be discarded.*

### ========== Question 2.4 --- [10 marks] ==========

Now we want to make clear how these eigenvectors and the mean are used to approximate the data. `pca.transform(digits.data[idx,:].reshape(1, -1)).flatten()` will transform the digit in row `idx` into pca space. Another way to think about this is that it will give you the coefficents to multiply each eigenvector by and to add to the mean digit such that you can reconstruct the digit.

For the digit with idx = 0 (which should itself be a zero), create 4 plots:
1. The original digit
1. The digit reconstructed using 1 principal component
1. The digit reconstructed using 2 principal components
1. The digit reconstructed using 5 principal components

In the plot titles show:
1. the number of principal components used
1. the percentage of variance explained by that number of principal components
1. the coefficients of each principal component rounded to nearest integer (tip, convert to integers to reduce print space), i.e. the PCA space vector.

Below the plots, comment on the result. Do the eigenvectors produce realistic data?

```python
# Plot the original digit.
orig = digits.data[0,:].reshape(8,8)
plt.title("The original digit")
plt.show(sns.heatmap(orig, xticklabels = np.arange(1,9,1), yticklabels = np.arange(1,9,1), cmap="YlGnBu"))

# Digit reconstruction
num_comp = [1, 2, 5]
for n in num_comp:
    
    pca = PCA(n_components=n, random_state = 100).fit(digits.data)
    mean_digit = pca.mean_.reshape(8,8)
    pca_space = pca.transform(digits.data[0,:].reshape(1, -1)).flatten()
    
    print ("Reconstruction using {} principal components.".format(n))
    percentage = round(np.sum(ratio[:n], axis = 0) * 100, 1)
    component_coefficients = pca_space[:n]
    eigenvectors = pca.components_[:n]
    
    reconstruction = (np.sum(eigenvectors * pca_space[:, np.newaxis], axis = 0)).reshape(8,8) + mean_digit
    title = "Reconstruction.\n"
    title += "Number of principal components: " + str(n) + ".\n"
    title += "Percentange of variance explained: " + str(percentage) + "%.\n"
    title += "Principal component coefficients: "
    for nn in range (n):
        if nn < n  - 1: title += str(int(component_coefficients[nn]))+", "
        else: title += str(int(component_coefficients[nn])) + '.'
    plt.title(title)
    plt.show(sns.heatmap(reconstruction, xticklabels = np.arange(1,9,1), yticklabels = np.arange(1,9,1), cmap="YlGnBu"))
```

*Using a single eigenvector does not result in a realistic plot, which is not surprising, since only 14.9% of variance is explained by it. However, increasing the amount of eigenvectors to 2 or 5 greatly increased the quality of the results. Using 2 eigenvectors, a somewhat realistic plot with vague but visible boundaries of number 0 was obtained. Using 5, the number 0 can be clearly seen in the construction, with contrasting colours (values) between the edges of the number and empty spaces around/within. A clear tendency of better results with higher number of components (eigenvectors) used was observed, which was also true for percentage of explained variance (1 component - 14.9%, 2 components: 28.5%, 5 components: 54.5%). As seen in question 2.2, the explained variance percentage does not increase in a stable manner, each addition of a new component (eigenvector) increases the percentage by less than the previous additions.*

*Adding 1 component when have:*
* *0, adds +14.9% explained variance.*
* *1, adds +13.6% explained variance.*

*Adding 3 components when already have 2 adds only +26% explained variance.*

Fit 4 models:
1. SVC with a linear kernel, no shrinking, and a random_state on digit data
1. SVC with a linear kernel, no shrinking, and the same random_state on pca transformed digit data (use the full 64 component pca above)
1. Gaussian Naive Bayes Classifier on digit data
1. Gaussian Naive Bayes Classifier on pca transformed digit data (use the full 64 component pca above)

Use 5 fold cross validation and take the mean fold score as the result. Plot or print the results.

Below the code, explain why one classifier improved when we used PCA, but the other did not!

```python
X = digits.data
y = digits.target
Xpca = pca.transform(X)

svc = np.mean(cross_val_score(SVC(kernel='linear', random_state = 0), X, y, cv=5))
svc_pca = np.mean(cross_val_score(SVC(kernel='linear', random_state = 0), Xpca, y, cv=5))
nb = np.mean(cross_val_score(GaussianNB(), X, y, cv=5))
nb_pca = np.mean(cross_val_score(GaussianNB(), Xpca, y, cv=5))

print ("The mean 5-fold cross validation scores: ")
print ("SVC                 : {}%".format(round(svc*100, 2)))
print ("SVC with PCA        : {}%".format(round(svc_pca*100, 2)))
print ("Gaussian NB         : {}%".format(round(nb*100, 2)))
print ("Gaussian NB with PCA: {}%".format(round(nb_pca*100, 2)))
```

*First of all, SVC tend to be robust even when using data that might span span only a subspace of full feature dimensions (which is the case with MNIST). SVC result did not improve, because it operates at sample level. SVC kernel computation is not calculated between the features, but at samples instead, making PCA useless. Furthermore, SVC tend to be robust even when using data that might span span only a subspace of full feature dimensions (which is the case with MNIST), which is often not the case for other classifiers, such as Gaussian NB. To be specific, when using Gaussian NB, it is crucial to have a very large set of samples if the feature space is large. Having a sample that is non-representative of the population will result in bad performance of the model, and having many features makes it very hard to obtain a representative sample. PCA allows to solve this problem by removing redundant features and making the dimensionality smaller, consequently improving the performance of the classifier.*
